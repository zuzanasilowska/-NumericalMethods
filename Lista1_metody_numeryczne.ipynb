Rozważamy dwa przybliżenia Padé funkcji \( e^{-x} \):

$$
z_1 = \frac{6 - 2x}{6 + 4x + x^2}
$$
$$
z_2 = \frac{6 - 4x + x^2}{6 + 2x}
$$

### Zadanie polega na:
1. Sporządzeniu wykresu przybliżeń Padé (na jednym rysunku).
2. Sporządzeniu tych samych przybliżeń na oddzielnych rysunkach w trybie wielowykresowym.
3. Sporządzeniu wykresów błędu każdego przybliżenia w porównaniu do funkcji \( e^{-x} \).

Do wykresów należy dodać tytuł, opisy osi oraz legendę. Należy również eksperymentować ze stylami, grubością i kolorami linii. Na końcu należy zapisać wykresy do formatu PNG i PDF.



<div style="text-align; font-family: 'Georgia'; font-size: 12px;">
    <h2> &#8618; Podpunkt pierwszy</h2>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Definicja funkcji z1 - pierwsze alternatywne przybliżenie Padégo 
def z1(x):
    return (6-2*x)/(6+4*x+x**2)

# Definicja funkcji z2 - drugie alternatywne przybliżenie Padégo 
def z2(x):
    return (6-4*x+x**2)/(6+2*x)

#oryginalna funkcja wykładnicza
def funkcja(x):
    return np.exp(-x)

x = np.linspace(-2,3, 1000)

wartości_1z= z1(x)
wartości_2z = z2(x)
wartości_funkcji=funkcja(x)


#tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.plot(x, wartości_1z, label="przybliżenie z1", color="darkcyan", linewidth=1.3)
plt.plot(x, wartości_2z, label="przybliżenie z2", color="plum", linewidth=1.3)
plt.plot(x, wartości_funkcji, label="funkcja exp(-x)", color="gold", linewidth=1.3)
plt.title("Przybliżenie Padégo funkcji exp(-x)", fontsize=12, fontweight='bold', color="dimgray")
plt.xlabel("x",fontsize=10, fontweight='bold', color="dimgray")
plt.ylabel("wartość funkcji",fontsize=10, fontweight='bold', color="dimgray")
plt.gca().set_facecolor('snow')
plt.legend()
plt.grid(True)
#zapisywanie wykres jako PNG na pulpicie
plt.savefig(os.path.join(os.path.expanduser("~"), "Desktop", 'wykres_przybliżenie_padego.png'), format='png', dpi=300)



plt.show()



```

Napisz program znajdujący dokładność maszynową na Twoim komputerze:   
W internecie epsilon maszynowy posiada dwie definicje i możemy dostosować kod w zależności od przyjętej.

<div style="text-align; font-family: 'Georgia'; font-size: 12px;">
    <h2>Definicja 1</h2>
</div>    
&#8618; Dokładność maszynowa to najmniejsza liczba, która może zostać dodana do liczby 
1, aby wynik był różny od 1.  

<div style="text-align; font-family: 'Georgia'; font-size: 12px;">
    <h2>Definicja 2</h2>
</div>    
&#8618; Jest to największa liczba nieujemna, której dodanie do jedności daje wynik równy 1.

Poniższy kod zaimplementowany jest dla definicji pierwszej. Kod dla definicji drugiej rożniłby się tym, że w pętli zwrócilibyśmy końcowo sam epsilon zamiast epsilon * 2.


```python
#kod dla pierwszej definicji
def dokladnosc_maszynowa():
    epsilon = 1.0
    while (1.0 + epsilon) != 1.0:  
        epsilon /= 2.0             
    return epsilon * 2.0           

dokladnosc = dokladnosc_maszynowa()
print(f"Dokładność maszynowa: {dokladnosc}")
```

```python
#kod dla drugiej definicji
def dokladnosc_maszynowa():
    epsilon = 1.0
    while (1.0 + epsilon) != 1.0:  
        epsilon /= 2.0             
    return epsilon        

dokladnosc = dokladnosc_maszynowa()
print(f"Dokładność maszynowa: {dokladnosc}")
```

Błąd zaokrąglenia występuje w obliczeniach komputerowych z powodu ograniczeń w sposobie, w jaki komputery przechowują liczby zmiennoprzecinkowe. Wykorzystują one system binarny do reprezentowania liczb, co oznacza, że liczby są zapisywane jako sekwencje zer i jedynek. Jednak nie wszystkie liczby dziesiętne mogą być dokładnie zapisane w systemie binarnym, podobnie jak niektóre liczby ułamkowe nie mogą być dokładnie zapisane w systemie dziesiętnym

Liczby zmiennoprzecinkowe są zapisywane w standardzie IEEE 754. Oznacza to, że tylko ograniczona liczba cyfr może zostać dokładnie zapisana. Gdy liczba ma więcej cyfr niż dostępna pamięć, nadmiar cyfr jest "obcinany", co powoduje zaokrąglenie liczby.

W przypadku wyrażenia 7/100 * 100 - 7 liczba 7/100 (czyli 0.07) nie jest dokładnie reprezentowana w systemie binarnym, co prowadzi do minimalnego błędu, który zostaje powielony przy dalszych operacjach. 

Liczby takie jak 7, 14, 28, 29 generują takie same błędy zaokrąglenia przy operacjach z ułamkami dziesiętnymi. Dzieje się tak, ponieważ ich reprezentacja w systemie binarnym wymaga przybliżenia (zaokrąglenia) na pewnym etapie.

<div style="text-align; font-family: 'Georgia'; font-size: 12px;">
    <h2>Zadanie 4</h2>
</div>     
Liczba 1,7 ma w standardzie IEEE przybliżoną reprezentację:

$$
\text{znak} \quad \text{wykładnik} \quad \text{część ułamkowa}
$$
$$
0 \quad 01111111 \quad 10110011001100110011001
$$

Jaki jest błąd bezwzględny i względny tego przybliżenia?



```python
# Reprezentacja binarna liczby 1.7 w standardzie IEEE 754:
# Wykładnik = 127 (01111111)
#Mantysa = 1.10110011001100110011001
# Znak = 0 

def przyblizona_wartosc():
    wykładnik = 127 - 127  # E - 127, ponieważ wykładnik to 127 w IEEE
    mantysa_bin = '1.10110011001100110011001'  # Mantysa w postaci binarnej
    mantysa = 1 + sum([int(bit) * 2**-(i+1) for i, bit in enumerate(mantysa_bin.split('.')[1])]) # Zamieniamy bity na liczby 0/1, i podnośimy do odpowiedniej potęgi 
    
    # Obliczamy wartość przybliżoną: mantysa * 2^wykładnik
    przyblizona = mantysa * 2**wykładnik
    return przyblizona

dokladna_liczba = 1.7

przyblizona_liczba = przyblizona_wartosc()

blad_bezwzgledny = abs(dokladna_liczba - przyblizona_liczba)

blad_wzgledny = blad_bezwzgledny / dokladna_liczba

print(f"Dokładna wartość: {dokladna_liczba}")
print(f"Przybliżona wartość: {przyblizona_liczba}")
print(f"Błąd bezwzględny: {blad_bezwzgledny}")
print(f"Błąd względny: {blad_wzgledny}")
```

    <h2>Zadanie 5</h2>
</div>     
Napisz funkcję obliczającą wartość wielomianu:

$$
W(x) = 6x^4 + 5x^3 - 13x^2 + x + 1
$$

dla zadanego x. Korzystając z niej, oblicz wartość wielomianu dla wszystkich x z przedziału:

$$
\langle -10, 10 \rangle
$$

przy kroku:

$$
dx = 0{,}0001
$$

Wykonaj profil tego programu.





```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Definiujemy funkcję wielomianową W(x), która zwraca wynik dla wielomianu 6x^4 + 5x^3 - 13x^2 + x + 1
def W(x):
    return 6*x**4 + 5*x**3 - 13*x**2 + x + 1

# Funkcja obliczenia() generuje wartości x oraz oblicza dla nich wartości wielomianu W(x)
def obliczenia():
    start_time = time.time() 
    x_wartosci = np.arange(-10, 10, 0.0001)
    w_wartosci = [W(x) for x in x_wartosci]
    end_time = time.time()  
    czas1 = end_time - start_time
    print(f"Czas wykonania obliczeń: {czas1} sekundy")
    return x_wartosci, w_wartosci, czas1

# Funkcja zapisywanie_do_pliku() zapisuje wyniki obliczeń do pliku tekstowego
def zapisywanie_do_pliku(x_wartosci, w_wartosci):
    with open("wielomian_wyniki1.txt", "w") as file:
        for x, w in zip(x_wartosci, w_wartosci):
            file.write(f'W({x:.4f}) = {w:.4f}\n')
            
#funkcja do rysowania wykresu funkcji wielomianowej
def rysuj_wykres(x_wartosci, w_wartosci):
    plt.figure(figsize=(7, 4))
    plt.plot(x_wartosci, w_wartosci, label="W(x) = 6x^4 + 5x^3 - 13x^2 + x + 1", color="plum")
    plt.title("Wykres funkcji wielomianowej W(x)")
    plt.xlabel("x")
    plt.ylabel("W(x)")
    plt.legend()
    plt.gca().set_facecolor('snow')
    plt.show()

x_wartosci, w_wartosci, czas1 = obliczenia()
zapisywanie_do_pliku(x_wartosci, w_wartosci)
rysuj_wykres(x_wartosci, w_wartosci)

```

</div>   
Za najbardziej ekonomiczny sposób wyliczania wartości wielomianu uznawany jest schemat Hornera. Korzystając z tego schematu, zmodyfikuj program z poprzedniego zadania. Ponownie wykonaj profil programu. Porównaj go z wcześniejszym profilem.



```python
import numpy as np  
import time  

# Funkcja W_horner(x_wartosci) implementuje schemat Hornera do obliczenia wartości wielomianu

def W_horner(x_wartosci):
    a = np.array([6, 5, -13, 1, 1])  # Współczynniki wielomianu: 6x^4 + 5x^3 - 13x^2 + x + 1
    result = a[0] * np.ones_like(x_wartosci)  
    # Schemat Hornera: iteracyjnie przetwarzamy każdy współczynnik
    for coefficient in a[1:]:
        result = result * x_wartosci + coefficient  # Aktualizujemy wynik zgodnie z algorytmem Hornera
    return result  

# Funkcja obliczenia_horner() mierzy czas obliczeń dla schematu Hornera
def obliczenia_horner():
    start_time = time.time()  
    x_wartosci = np.arange(-10, 10, 0.0001) 
    w_wartosci = W_horner(x_wartosci)  # Obliczamy wartości wielomianu W(x) schematem Hornera
    end_time = time.time()  
    czas2 = end_time - start_time  

    # Wyświetlamy czas wykonania obliczeń
    print(f"Czas wykonania działania sposobem schematu Hornera: {czas2:.4f} sekundy")
    return x_wartosci, w_wartosci, czas2  

# Funkcja zapisywanie_do_pliku_horner() zapisuje wyniki obliczeń do pliku tekstowego
def zapisywanie_do_pliku_horner(x_wartosci, w_wartosci):
    with open("wielomian_wyniki_horner_vec.txt", "w") as file:
        for x, w in zip(x_wartosci, w_wartosci):
            file.write(f'W({x:.4f}) = {w:.4f}\n')  

# Funkcja rysuj_wykres_horner() rysuje wykres funkcji wielomianowej W(x) obliczonej schematem Hornera
def rysuj_wykres_horner(x_wartosci, w_wartosci):
    plt.figure(figsize=(7, 4))  
    plt.plot(x_wartosci, w_wartosci, label="W(x) (schemat Hornera)", color="plum")
    plt.title("Wykres funkcji wielomianowej W(x) obliczony schematem Hornera")
    plt.xlabel("x")
    plt.ylabel("W(x)")
    plt.gca().set_facecolor('snow') 
    plt.legend()
    plt.show()  


x_wartosci, w_wartosci, czas2 = obliczenia_horner()


zapisywanie_do_pliku_horner(x_wartosci, w_wartosci)

rysuj_wykres_horner(x_wartosci, w_wartosci)
```
